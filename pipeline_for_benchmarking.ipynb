**To recreate the time/memory usage figure as seen in the paper**

While skandiver can be used either with a directory of FASTA files or a singular FASTA file, geNomad and MobileElementFinder (MEFinder) must be run on a singular FASTA file input. The script merge_fna.py provided in the repository can be used to merge all the provided benchmark assemblies into a single file for use.
<br>
<br>

To obtain wall time and maximum memory usage metrics for each software, simply run the following commands:

**skandiver Figure 6a**<br>
Important note for skandiver: the most recent scripts used in skandiver have been updated to better support query genomes where the species of origin is unknown; because of this, runtimes will be slightly higher with the latest versions of skandiver.
```
/usr/bin/time -v ./skandiver.sh <path_to_fig6a_data> fig6a_skandiver 10000 <path_to_database>
```

For instance, if you followed the tutorial for setting up the GTDB database for comparison against (https://github.com/YoukaiFromAccounting/skandiver), the command could look something like:
```
/usr/bin/time -v ./skandiver.sh data/fig6a/CP129688.1_random.fasta fig6a_skandiver 10000 gtdb_skani_database_ani
```

**skandiver Figure 6b**
```
/usr/bin/time -v ./skandiver.sh <path_to_fig6b_data> fig6b_skandiver 10000 <path_to_database>
```
<br>
<br>

**geNomad Figure 6a**
```
/usr/bin/time -v genomad end-to-end --cleanup --splits 8 <path_to_fig6a_data> fig6a_GMD genomad_db
```

**geNomad Figure 6b**
```
/usr/bin/time -v genomad end-to-end --cleanup --splits 8 <path_to_fig6b_data> fig6b_GMD genomad_db
```
<br>
<br>

**MobileElementFinder Figure 6a**<br>
Important note for MEFinder: if you are receiving a "Permission Denied" error for the files you are trying to access, a workaround is to create a new directory **tmp** and add **--temp-dir tmp** to the end of your mefinder command.
```
/usr/bin/time -v mefinder find --contig <path_to_fig6a_data> fig6a_MEF
```

**MobileElementFinder Figure 6b**
```
/usr/bin/time -v mefinder find --contig <path_to_fig6b_data> fig6b_MEF
```
<br>
