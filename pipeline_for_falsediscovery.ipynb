**To recreate the conserved element control table as seen in the paper**

In Table 1, skandiver, geNomad, and MobileElementFinder were each run on a set of approximately two thousand known conserved genes and elements, with the goal of determining the false positive rate of each of the software (ie: how many of these elements would be falsely identified as a mobile element). Table 1 lists the five conserved elements larger than 10kbp in length present in this set, for which neither skandiver, geNomad, nor MobileElementFinder identified as a mobile element.

To reproduce this table, simply run skandiver, geNomad, and MobileElementFinder on the complete set of conserved elements and observe the outputs of the softwares. For the five conserved elements larger than 10kbp, the outputs should look like the following:
